```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
LOG = 'workspace_boundary_gen.csv'
my_data = np.genfromtxt(LOG, delimiter=',')
upper = my_data[:,0]
lower_max = my_data[:,1]
lower_min = my_data[:,2]
```

```python
fig = plt.figure(figsize = (20, 15))

plt.plot(upper, lower_max, linewidth=3)
plt.plot(upper, lower_min, linewidth=3)
plt.plot()


plt.xlabel('Upper angle (degrees)', fontsize = 30)
plt.ylabel('Lower angle (degrees)', fontsize = 30)
plt.title('Workspace Boundary', fontsize = 30)
plt.legend(['Maximum angle', 'Minimum angle'], fontsize = 25)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.show()
```
