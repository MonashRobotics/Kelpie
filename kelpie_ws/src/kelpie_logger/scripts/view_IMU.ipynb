```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
LOG = '20241014-185228.csv'
my_data = np.genfromtxt('../logs/' + LOG, delimiter=',')

time_data = my_data[:, 0]
angle_data = np.rad2deg(my_data[:, 1:])
```

```python
idxs = np.logical_and(time_data > 39, time_data < 50)

fig = plt.figure(figsize = (20, 15))
plt.plot(time_data, angle_data[:, 0])


plt.xlabel('time (s)')
plt.ylabel('Angle (Deg)')
plt.title('Roll')
plt.show()
```

```python

idxs = np.logical_and(time_data > 39, time_data < 50)

fig = plt.figure(figsize = (20, 15))
plt.plot(time_data[idxs], angle_data[idxs, 0])


plt.xlabel('time (s)')
plt.ylabel('Angle (Deg)')
plt.title('Roll')
plt.show()
```
